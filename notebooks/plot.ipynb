- riusciamo a fittare? nel caso P=6000, consideriamo solo il periodo iniziale di stabilit e crescita monotona dell'accuracy.
    - in tutti i casi le performance migliorano con H.
    - per P=60,600 riusciamo a ottenere un fit quasi perfetto con H abbastanza grande.
    - per P=6000, H=1600 ancora non basta a fittare (86% max train acc).
